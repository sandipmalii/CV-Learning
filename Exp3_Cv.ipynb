```python
# Install required libraries
!pip install scipy scikit-image gdown

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io
from skimage.util import random_noise
from skimage.filters import laplace
import gdown

# Download the image from Google Drive
file_id = '10uJrjMsw4c5sEsdPnlfdxGlN3CqpzFtk'
url = f'https://drive.google.com/uc?id={file_id}'
output = 'image.jpg'
gdown.download(url, output, quiet=False)

# Load and display the image
image = io.imread('image.jpg', as_gray=True)

# Add random noise to the image
noisy_image = random_noise(image, mode='s&p', amount=0.1)
noisy_image = np.array(255 * noisy_image, dtype='uint8')

# Display the noisy image
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title('Original Noisy Image')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

# Apply 2-D Laplacian High-Pass Filter
laplacian = laplace(noisy_image)
sharpened_image = noisy_image - laplacian
sharpened_image = np.clip(sharpened_image, 0, 255)

# Display the sharpened image
plt.subplot(1, 3, 2)
plt.title('Sharpened Image')
plt.imshow(sharpened_image, cmap='gray')
plt.axis('off')

# Display Laplacian filter result
plt.subplot(1, 3, 3)
plt.title('Laplacian Filter')
plt.imshow(laplacian, cmap='gray')
plt.axis('off')
plt.show()
```
