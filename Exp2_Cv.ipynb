```python
!pip install scipy scikit-image

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import median_filter
from skimage import data
from skimage.util import random_noise
from skimage.filters.rank import modal
from skimage.morphology import disk

# Load example image
image = data.camera()

# Add random noise to the image
noisy_image = random_noise(image, mode='s&p', amount=0.1)
noisy_image = np.array(255 * noisy_image, dtype='uint8')

# Display the noisy image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Original Noisy Image')
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

# Apply Median Filter
median_filtered_image = median_filter(noisy_image, size=3)

# Display the median filtered image
plt.subplot(1, 2, 2)
plt.title('Median Filtered Image')
plt.imshow(median_filtered_image, cmap='gray')
plt.axis('off')
plt.show()

# Apply Mode Filter (requires defining a structuring element)
footprint = disk(1)  # disk-shaped structuring element with radius 1
mode_filtered_image = modal(noisy_image, footprint)

# Display the mode filtered image
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Mode Filtered Image')
plt.imshow(mode_filtered_image, cmap='gray')
plt.axis('off')
plt.show()
```
